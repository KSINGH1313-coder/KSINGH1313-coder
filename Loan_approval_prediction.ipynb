```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('loan_approval_dataset.csv')
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head(1)
```

```python
not_req=[' education', ' self_employed', 'loan_id', ' no_of_dependents', ]
df=df.drop(not_req,axis=1,)
```

```python
df.head(1)
```

```python
df.columns
```

```python
df.head()
```

```python
df[' loan_status']=df[' loan_status'].map({' Approved':1,' Rejected':0})
```

```python
df.columns
```

```python
df.head()
```

```python
x=df.drop(' loan_status',axis=1)
y=df[' loan_status']
```

```python
x.head()
```

```python
y.head()
```

#Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
models=dict()
```

#Model Training

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_score=lr.score(x_test,y_test)
print(lr_score*100)
models['LogisticRegression']= lr_score*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf_score=rf.score(x_test,y_test)
print(rf_score*100)
models['RandomForestClassifier']= rf_score*100
```

```python

from xgboost import XGBClassifier

xgb = XGBClassifier(scale_pos_weight=0.61, eval_metric='logloss')
xgb.fit(x_train, y_train)
```

```python
xg_score=xgb.score(x_test,y_test)
print(xg_score*100)
models['XGBClassifier']= xg_score*100
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt_score=dt.score(x_test,y_test)
print(dt_score*100)
models['DecisionTreeClassifier']= dt_score*100
```

#Model Results

```python
models
```

```python
model_names=list(models.keys())
test_scores=list(map(float,models.values()))
plt.bar(model_names,test_scores)
```
